```python
#IMPORTING THE LIBRARIES
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

#IGNORE ANY WARNINGS THAT MAY COME UP
import warnings
warnings.filterwarnings('ignore')
```

```python
#IMPORTING THE DATASET
mental_health_data = pd.read_csv("C://Users/FCSE20-011/Documents/AI ASSIGNMENT/CODE/students_mental_health_survey.csv")
mental_health_data
```

```python
#RENAMING COLUMN
mental_health_data1 = mental_health_data.rename(columns= {"Counseling_Service_Use": "Counselling"})
mental_health_data1
```

```python
#ENCODING CATEGORIACAL VARIABLES WITH LABEL ENCODING
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
#THE COLUMNS THAT NEED TO BE CONVERTED 
le = ["Course", "Gender", "Sleep_Quality", "Physical_Activity", "Diet_Quality", "Social_Support","Relationship_Status",
      "Substance_Use","Counselling","Extracurricular_Involvement","Family_History","Chronic_Illness","Residence_Type"]

#APPLYING THE ENCODING TO THE SPECIFIED COLUMNS ABOVE
for column in le:
    mental_health_data1[column] = label_encoder.fit_transform(mental_health_data1[column])
    

```

```python
#ADDING A COLUMN AND DISPLAYING THE DATASET
np.random.seed(42)
mental_health_data1['Mental_Health'] = np.random.choice([0, 1], size=len(mental_health_data1))
mental_health_data1
```

```python
#USING CORRELATION MATRIX ALONG WITH A HEATMAP TO GET A CLEAR UNDERSTANDING OF HOW MY VARIABLES RELATE
corr_matrix = mental_health_data1.corr()
plt.figure(figsize=(20, 20))#FIGURE OF THE GRAPH
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Variables")
plt.show()
```

```python
# Specify the column to use as a criterion for selecting the best columns
criterion_column = 'Mental_Health'

# Get the absolute correlation values of each feature with the criterion column
correlation_with_criterion = corr_matrix[criterion_column].abs()

# Select a threshold for correlation strength
correlation_threshold = 0.01  # Adjust this threshold as needed

# Select columns that have a correlation above the threshold
selected_columns = correlation_with_criterion[correlation_with_criterion > correlation_threshold].index

# Remove columns that are not selected
df_selected = mental_health_data1[selected_columns]

# Display the selected DataFrame
print("Selected Columns:")
print(df_selected.head())
```

```python
#REMOVING  NAN VALUES AND CHECKING IF THERE ARE REMOVED
mental_health_data3= df_selected.dropna()
mental_health_data3
```

```python
#RECHECKING IF THE NAN VALUES ARE STILL PRESENT
mental_health_data3.isna().any()
```

```python
#SPLITING THE DATASET INTO FEATURES 
X = mental_health_data3.drop('Mental_Health',axis= 1)
y = mental_health_data3['Mental_Health']
```

```python
#SPLITING DATA INTO TRAINING AND TEST SETS

#IMPORTING LIBRARIES
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.33)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.model_selection import cross_val_score
#RNADOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)

# CREATING AND TRIANING THE MODEL
clf = RandomForestClassifier(n_estimators=50, random_state=42)
clf.fit(X_train, y_train)

#MAKING PREDICTIONS ON THE TEST DATA
y_pred = clf.predict(X_test)

#CALCULATING ACCURACY AND CONFUSION MATRIX TO TEST THE MODEL
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

#PERFORMING CROSS-VALIDATION
cv_scores_clf = cross_val_score(clf, X_train, y_train, cv=10)
print("Cross-Validation Scores:")
print(cv_scores_clf)
print("Mean Cross-Validation Score:", cv_scores_clf.mean())

#classification report
print("Classification report:")
print(classification_report(y_pred, y_test))



```

```python
#DECISION TREE CLASSIFICATION

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)

#CREATING AND TRAINING THE MODEL 
dtc = DecisionTreeClassifier()
# dtc.fit(X_train, y_train)

#USING HYPERPARAMETER TUNING TO BOOST THE PERFOMANCE USING GRID SEARCH
param_grid = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(dtc, param_grid, cv=5)
grid_search.fit(X_train, y_train)

dtc_best = grid_search.best_estimator_

#MAKING PREDICTIONS ON THE TEST DATA
y_pred = dtc_best.predict(X_test)

#CALCULATING ACCURACY AND CONFUSION MATRIX TO TEST THE MODEL 
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)

#PERFORMING CROSS-VALIDATION
cv_scores_dtc = cross_val_score(dtc_best, X_train, y_train, cv=10)
print("Cross-Validation Scores:")
print(cv_scores_dtc)
print("Mean Cross-Validation Score:", cv_scores_dtc.mean())

#CLASSSIFICATION REPORT
print("Classification report:")
print(classification_report(y_test, y_pred))
```
