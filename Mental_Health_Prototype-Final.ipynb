```python
#ENCODING CATEGORIACAL VARIABLES WITH LABEL ENCODING
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
#THE COLUMNS THAT NEED TO BE CONVERTED 
le = ["Course", "Gender", "Sleep_Quality", "Physical_Activity", "Diet_Quality", "Social_Support",
      "Substance_Use"]

#APPLYING THE ENCODING TO THE SPECIFIED COLUMNS ABOVE
for column in le:
    mental_health_data2[column] = label_encoder.fit_transform(mental_health_data2[column])
    

import pickle
#SAVING THE LABEL ENCODERS AS A PICKLED FILE
with open('label_encoder.pkl', 'wb') as file:
    pickle.dump(label_encoder, file)

#DISPLAYING A FEW ROWS OF THE DATASET     
mental_health_data2.head() 
#DISPLAYING THE DATASET
mental_health_data2
```

```python
#ADDING A COLUMN AND DISPLAYONG THE DATASET
np.random.seed(42)
mental_health_data2['Mental_Health'] = np.random.choice([0, 1], size=len(mental_health_data2))
mental_health_data2
```

```python
#CHECKING THE VARIABLE CONTENTS OF THE COLUMN AND CHECKING FOR THE AMOUNT EACH NUMERICAL VALUE HAS IN THIS CASE 0 AND 1 
print("Uniques:",mental_health_data2['Mental_Health'].unique())

print("Count:",mental_health_data2['Mental_Health'].value_counts())
```

```python
#ELIMINATING ANY OCURRNECES OF ZEROS IN SPECIFIED COLUMNS
eliminate_zero = ['Stress_Level', 'Depression_Score', 'Anxiety_Score','Financial_Stress']

for column in eliminate_zero:
    #REPLACES OCCURENCES OF ZEROS WITH NAN VALUES
    mental_health_data2[column] = mental_health_data2[column].replace(0, np.NaN)
    
    #CALCULATES THE MEAN INCLUDING NAN VALUES AND CONVERTS IT TO AN INNTEGER
    mean = int(mental_health_data2[column].mean(skipna=True))
    
    #REPLACES THE NAN VALUES WITH THE CALCULATED MEAN
    mental_health_data2[column] = mental_health_data2[column].replace(np.NaN, mean) 
```

```python
#RECHECKING FOR DATA IN ALL COLUMNS
mental_health_data2.isna().any()
```

```python
#DISPLAYING THE FIRST FEW ROWS AGAIN
mental_health_data2.head()
```

```python
#FEATURE ENGINEERING BY COMBINING COLUMNS

#ROUNDING THE ADDITION AND DEVEISION OF COLUMS WITH THE NUMBER OF ROWS TO ONE DECIMAL PLACE
mental_health_data2['Emotional_Distress']= round((mental_health_data2['Stress_Level'] + mental_health_data2['Depression_Score'] + mental_health_data2['Anxiety_Score']) / 3)
```

```python
# #HANDLING OUTLIERS BASED ON THE BOXPLOT RESULTS OF AGE

#FINDING THE UPPER AND LOWER LIMITS OF AGE USING THE MEAN AND STANDARD DEVIATION
upper_limit = mental_health_data4['Age'].mean() + 3* mental_health_data4['Age'].std()
lower_limit = mental_health_data4['Age'].mean() - 3* mental_health_data4['Age'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
#FINDING THE OUTLIERS
mental_health_data4.loc[(mental_health_data4['Age'] > upper_limit) | (mental_health_data4['Age'] < lower_limit)]
```

```python
#DELETING THE OULIER DATA
mental_health_data5= mental_health_data4.loc[(mental_health_data4['Age'] <= upper_limit) & (mental_health_data4['Age'] >= lower_limit)]
print('before removing outliers:', len(mental_health_data4))
print('after removing outliers:',len(mental_health_data5))
print('outliers:', len(mental_health_data4)-len(mental_health_data5))
```

```python
#SHOWING THE GRAPH AGAIN
sns.boxplot(mental_health_data5['Age'])
```

```python
#CHANGING THE OULIER VALUES TO UPPER AND LOWER LIMIT VALUES
mental_health_data6 = mental_health_data5.copy()
mental_health_data6.loc[(mental_health_data6['Age']>upper_limit), 'Age'] = upper_limit
mental_health_data6.loc[(mental_health_data6['Age']<lower_limit), 'Age'] = lower_limit
```

```python
#SPLITING DATA INTO TRAINING AND TEST SETS

#IMPORTING LIBRARIES
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.33)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
#DECISION TREE CLASSIFICATION

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)

#CREATING AND TRAINING THE MODEL 
dtc = DecisionTreeClassifier()
# dtc.fit(X_train, y_train)

#USING HYPERPARAMETER TUNING TO BOOST THE PERFOMANCE USING GRID SEARCH
param_grid = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(dtc, param_grid, cv=5)
grid_search.fit(X_train, y_train)

dtc_best = grid_search.best_estimator_

#MAKING PREDICTIONS ON THE TEST DATA
y_pred = dtc_best.predict(X_test)

#CALCULATING ACCURACY AND CONFUSION MATRIX TO TEST THE MODEL 
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)

#PERFORMING CROSS-VALIDATION
cv_scores_dtc = cross_val_score(dtc_best, X_train, y_train, cv=10)
print("Cross-Validation Scores:")
print(cv_scores_dtc)
print("Mean Cross-Validation Score:", cv_scores_dtc.mean())

#CLASSSIFICATION REPORT
print("Classification report:")
print(classification_report(y_test, y_pred))
```

```python
#K NEAREST NEIGHBORS CLASSIFICATION
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix, f1_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)

# CREATING AND TRIANING THE MODEL
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
knnc = KNeighborsClassifier()
knnc.fit(X_train, y_train)

#MAKING PREDICTIONS ON THE TEST DATA
y_pred = knnc.predict(X_test)

# CALCULATING ACCURACY,F1 SCORE AND CONFUSION MATRIX TO TEST THE MODEL 
fscore = f1_score(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"F1 score: {fscore}")
print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)

#PERFORMING CROSS-VALIDATION
cv_scores_knnc = cross_val_score(knnc, X_train, y_train, cv=10)
print("Cross-Validation Scores:")
print(cv_scores_knnc)
print("Mean Cross-Validation Score:", cv_scores_knnc.mean())

#CLASSIFICATION REPORT
print("Classification report:")
print(classification_report(y_pred, y_test))
```

```python
#LOGISTIC REGRESSION 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)
      
# CREATING AND TRIANING THE MODEL
lr = LogisticRegression()
lr.fit(X_train, y_train)

#MAKING PREDICTIONS ON THE TEST DATA
y_pred = lr.predict(X_test)

#CALCULATING ACCURACY AND CONFUSION MATRIX TO TEST THE MODEL
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)

#PERFORMING CROSS-VALIDATION
cv_scores_lr = cross_val_score(lr, X_train, y_train, cv=10)
print("Cross-Validation Scores:")
print(cv_scores_lr)
print("Mean Cross-Validation Score:", cv_scores_lr.mean())


#CLASSIFICATION REPORT
print("Classification report:")
print(classification_report(y_pred, y_test))
```

```python
#RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)

# CREATING AND TRIANING THE MODEL
rfc = RandomForestClassifier(n_estimators=50, random_state=42)
rfc.fit(X_train, y_train)

#MAKING PREDICTIONS ON THE TEST DATA
y_pred = rfc.predict(X_test)

#CALCULATING ACCURACY AND CONFUSION MATRIX TO TEST THE MODEL
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

#PERFORMING CROSS-VALIDATION
cv_scores_rfc = cross_val_score(rfc, X_train, y_train, cv=10)
print("Cross-Validation Scores:")
print(cv_scores_rfc)
print("Mean Cross-Validation Score:", cv_scores_rfc.mean())

#classification report
print("Classification report:")
print(classification_report(y_pred, y_test))
```
